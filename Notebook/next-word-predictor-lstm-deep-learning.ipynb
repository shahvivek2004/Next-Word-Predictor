```python
#from sklearn.model_selection import train_test_split
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding,LSTM,Dense

print("DONE")
```

```python
def padding(seq):
  #max_len=max([len(x) for x in seq])
  #print(max_len)
  seq=pad_sequences(seq,padding='pre',maxlen=86)
  return seq
```

```python
# dataExtraction

text=open("/kaggle/input/next-word-predictor-data/dataset.txt").read().split('.')
#print(text)
```

```python
# dataPreprocessing

tk=Tokenizer()
tk.fit_on_texts(text)
seq=tk.texts_to_sequences(text)
# print(seq)
```

```python
#print(tk.word_index)
```

```python
#logicBuilding

x=[]
y=[]

for i in seq:
  for j in range(0,len(i)-1):
    # print(f"{i[:j+1]} and {i[j+1]}")
    x.append(i[:j+1])
    y.append(i[j+1])

x=padding(x)
y=np.array(y)
x=np.array(x)
#in order to perform task either we should go towards regression or classification
#now if we use regression then on fractional values it gives "no result" , so better we should go towards multiclass classification
#One Hot Encoding (OHE):
no_unique_word=len(tk.word_index)
y=to_categorical(y,num_classes=no_unique_word+1)
#x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
print(x)
print(y)
```

```python
print(no_unique_word)
```

```python
#modelBuilding
model=Sequential()
model.add(Embedding(input_dim=no_unique_word+1,output_dim=100, input_length=len(x[0])))
model.add(LSTM(150))
model.add(Dense(no_unique_word+1,activation='softmax'))
model.summary()
```

```python
model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
model.fit(x,y,epochs=100)
```

```python
loss, accuracy = model.evaluate(x, y)
print(f'Train Loss: {loss}, Train Accuracy: {accuracy}')
```
